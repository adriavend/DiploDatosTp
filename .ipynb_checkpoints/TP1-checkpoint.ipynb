```python
import matplotlib.pyplot as plt
import numpy
import pandas
import seaborn
```

```python
seaborn.__version__
```

```python
dataset = pandas.read_csv('./datasets/hfi_cc_2018.csv')
dataset.shape
```

```python
dataset.columns
```

```python
important_cols = ['year', 'ISO_code', 'countries', 'region']

important_cols += [col for col in dataset.columns if 'pf_identity' in col]

important_cols += [
    'ef_score', # Economic Freedom (score)
    'ef_rank', # Economic Freedom (rank)
    'hf_score', # Human Freedom (score)
    'hf_rank', # Human Freedom (rank)
]

dataset_numeric = dataset[important_cols].select_dtypes(include=['number'])
```

```python
dataset[important_cols]
```

# 1.1 - Rango de variables

```python
for c in important_cols:
    print(c + ': ' + str(dataset[c].min()) + ' - ' + str(dataset[c].max()))
```

# 1.2 Obtener media, mediana y desviación estándar de las variables pf_identity y hf_score en el mundo y compararla con la de Latinoamérica y el caribe. ¿Tiene sentido calcular la moda?

```python


dataset_LAC = dataset.loc[dataset['region'] == 'Latin America & the Caribbean']

print('Media de pf_identity: MUNDIAL', dataset['pf_identity'].mean(), 'vs', 'LA y Caribe', dataset_LAC['pf_identity'].mean())
print('Mediana de pf_identity: MUNDIAL', dataset['pf_identity'].median(), 'vs', 'LA y Caribe', dataset_LAC['pf_identity'].median())
print('Desviacion estandar de pf_identity: MUNDIAL', dataset['pf_identity'].std(), 'vs', 'LA y Caribe', dataset_LAC['pf_identity'].std())
print('---------------------------------------------------------')
print('Media de hf_score: MUNDIAL', dataset['hf_score'].mean(), 'vs', 'LA y Caribe', dataset_LAC['hf_score'].mean())
print('Mediana de hf_score: MUNDIAL', dataset['hf_score'].median(), 'vs', 'LA y Caribe', dataset_LAC['hf_score'].median())
print('Desviacion estandar de hf_score: MUNDIAL', dataset['hf_score'].std(), 'vs', 'LA y Caribe', dataset_LAC['hf_score'].std())

```

No tiene sentido calcular la moda ya que la misma al representar el valor mas repetido solo es aplicable para variables continuas

Son variables directamente comparables. Podría ser influenciadas por variables como pais-sexo-edad.

# 1.4 ¿Cómo pueden sanearse los valores faltantes?

Los valores faltantes pueden sanearse de maneras diferentes, podemos sanearlos usando la Media de los valores válidos. 
Sin embargo también podemos sanear los valores faltantes mediante la imputación por regresión, esto pretende de alguna manera estimar los valores faltantes en base a su relación con las otras variables mediante un analisis de regresion.

Nosotros decidimos subsanar los datos faltantes utilizando el método de sustitución por la Media asumiendo los riesgos que este método nos trae, sabiendo que dificulta la estimación de la variancia y distorsiona la verdadera distribución de la variable ya que añade valores constantes basadas en la media de nuestra variable con valores válidos.


```python
for col in dataset_numeric:
    dataset_numeric[col] = dataset_numeric[col].fillna(dataset_numeric[col].mean())
```

En este caso hemos elegido la media de toda la columna, pero podría haberse reemplazado para cada región por la media de cada región por ejemplo.

## 1.5 ¿Encuentra outliers en estas dos variables? ¿Qué método utiliza para detectarlos? ¿Los outliers, son globales o por grupo? ¿Los eliminaría del conjunto de datos?

Los outliers pueden ser globales y por grupo en el siguiente ejemplo muestra cómo podemos identificar outliers globales, ya que no le indicamos al gráfico que agrupe por un atributo específico. Por lo cual identifica los outliers de acuerdo en base a la media de la variable.

```python
dataset[important_cols].boxplot(column='pf_identity', figsize=(14,6))
plt.show()
```

Por otra parte podemos definir un grupo con la misma variable pero esta vez agrupando por Year, de esta manera podemos detectar los outliers de acuerdo al atributo que le estamos indicando que agrupe.

```python
dataset[important_cols].boxplot(column='pf_identity',by='year', figsize=(14,6))
plt.show()
```

# ¿Encuentra outliers en estas dos variables? 
Podemos notar valores que están alejados significativamente de la media de la variable, los cuales pueden ser considerados como outliers si los mismos no representan valores significativos para el análisis.

## ¿Qué método utiliza para detectarlos?
Para detectar los Outliers generamos un gráfico de tipo boxplot con el cual pretendemos distinguir con solo la observación de los datos aquellos valores ubicados en los extremos superiores e inferiores como los outliers, dejando como valores válidos los que se encuentran en medio de los cuartiles superior e inferior.


## 2.1 Grafiquen la media de la variable *pf_identity* y *hf_score* a través de los años.

```python
plt.figure(figsize=(10,6))
seaborn.barplot(data=dataset, x='year', y='pf_identity')
plt.ylim(5, 10)

print('Mitad del Rango pf_identity: ', (dataset['pf_identity'].min() + ((dataset['pf_identity'].max()-dataset['pf_identity'].min())/2)))

plt.figure(figsize=(10,6))
seaborn.barplot(data=dataset, x='year', y='hf_score')
plt.ylim(6.5, 9.12)

print('Mitad del Rango hf_score: ', (dataset['hf_score'].min() + ((dataset['hf_score'].max()-dataset['hf_score'].min())/2)))
```

Como en ambos casos los valores de la media supera el valor de la mitad del rango, se decide tomar éste como límite inferior de la escala para analizar por cuanto se sobrepasa la media de este valor intermedio. El límite inferior es el mayor valor posible que pueden tomar los datos.
